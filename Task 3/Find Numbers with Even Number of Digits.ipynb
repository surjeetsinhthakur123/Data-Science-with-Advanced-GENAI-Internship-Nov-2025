```python
# 1295. Find Numbers with Even Number of Digits

# Given an array nums of integers, return how many of them contain
# an even number of digits.

# Example 1:

# Input: nums = [12, 345, 2, 6, 7896]
# Output: 2

# Explanation:
# 12 contains 2 digits (even number of digits).
# 345 contains 3 digits (odd number of digits).
# 2 contains 1 digit (odd number of digits).
# 6 contains 1 digit (odd number of digits).
# 7896 contains 4 digits (even number of digits).
# Therefore only 12 and 7896 contain an even number of digits.

# Example 2:

# Input: nums = [555, 901, 482, 1771]
# Output: 1

# Explanation:
# Only 1771 contains an even number of digits.

# Constraints:
# 1 <= nums.length <= 500
# 1 <= nums[i] <= 10^5
```

```python
# Define the Solution class
class Solution:
    # Define the function to count numbers with even number of digits
    def findNumbers(self, nums):
        # Initialize a counter to store count of valid numbers
        count = 0
        
        # Loop through each number in the nums array
        for num in nums:
            # Convert the number to string and find its length (number of digits)
            digits = len(str(num))
            
            # Check if the number of digits is even
            if digits % 2 == 0:
                # If even, increase the count
                count += 1
        
        # Return the final count
        return count
```

```python
nums = [12, 345, 2, 6, 7896]
obj = Solution()
print(obj.findNumbers(nums))
```
