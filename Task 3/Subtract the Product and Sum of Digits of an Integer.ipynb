```python
# 1281. Subtract the Product and Sum of Digits of an Integer

# Given an integer number n,
# return the difference between the product of its digits
# and the sum of its digits.

# Example 1:
# Input: n = 234
# Output: 15
# Explanation:
# Product of digits = 2 * 3 * 4 = 24
# Sum of digits = 2 + 3 + 4 = 9
# Result = 24 - 9 = 15

# Example 2:
# Input: n = 4421
# Output: 21
# Explanation:
# Product of digits = 4 * 4 * 2 * 1 = 32
# Sum of digits = 4 + 4 + 2 + 1 = 11
# Result = 32 - 11 = 21

# Constraints:
# 1 <= n <= 10^5
```

```python
class Solution:
    def subtractProductAndSum(self, n: int) -> int:
        product = 1        # to store product of digits
        summation = 0      # to store sum of digits
        
        while n > 0:                       # loop until all digits are processed
            digit = n % 10                 # get the last digit
            product = product * digit     # multiply digit to product
            summation = summation + digit # add digit to sum
            n = n // 10                   # remove the last digit
        
        return product - summation         # return product minus sum
```

```python
sol = Solution()
print(sol.subtractProductAndSum(234))  
print(sol.subtractProductAndSum(4421))  
```
