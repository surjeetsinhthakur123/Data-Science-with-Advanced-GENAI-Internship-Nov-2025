```python
# 1486. XOR Operation in an Array

# You are given an integer n and an integer start.

# Define an array nums where:
# nums[i] = start + 2 * i (0-indexed)
# and n == nums.length.

# Return the bitwise XOR of all elements of nums.

# Example 1:
# Input: n = 5, start = 0
# Output: 8
# Explanation:
# Array nums is equal to [0, 2, 4, 6, 8]
# where (0 ^ 2 ^ 4 ^ 6 ^ 8) = 8.
# Where "^" corresponds to bitwise XOR operator.

# Example 2:
# Input: n = 4, start = 3
# Output: 8
# Explanation:
# Array num
```

```python
class Solution:
    def xorOperation(self, n: int, start: int) -> int:
        # Initialize result to 0 (identity for XOR)
        result = 0
        
        # Loop through indices from 0 to n-1
        for i in range(n):
            # Calculate the current value as start + 2*i
            num = start + 2 * i
            
            # Apply XOR with the result
            result ^= num
        
        # Return the final XOR result
        return result
```

```python
sol = Solution()
print(sol.xorOperation(5, 0))
```
