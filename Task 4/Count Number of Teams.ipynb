```python
# 1395. Count Number of Teams

# There are n soldiers standing in a line.
# Each soldier is assigned a unique rating value.

# You have to form a team of 3 soldiers amongst them under the following rules:

# • Choose 3 soldiers with index (i, j, k) with rating (rating[i], rating[j], rating[k]).
# • A team is valid if:
#   - rating[i] < rating[j] < rating[k]
#   - OR rating[i] > rating[j] > rating[k]
# • Where 0 <= i < j < k < n.

# Return the number of teams you can form given the conditions.
# Soldiers can be part of multiple teams.

# Example 1:
# Input: rating = [2,5,3,4,1]
# Output: 3
# Explanation:
# We can form three teams given the conditions:
# (2,3,4), (5,4,1), (5,3,1)

# Example 2:
# Input: rating = [2,1,3]
# Output: 0
# Explanation:
# We can't form any team given the conditions.

# Example 3:
# Input: rating = [1,2,3,4]
# Output: 4

# Constraints:
# • n == rating.length
# • 3 <= n <= 1000
# • 1 <= rating[i] <= 10^5
# • All the integers in rating are unique
```

```python
class Solution:
    def numTeams(self, rating):
        n = len(rating)          # total number of soldiers
        count = 0               # to count valid teams
        
        # choose middle soldier j
        for j in range(n):
            left_smaller = 0     # count of i < j where rating[i] < rating[j]
            left_greater = 0    # count of i < j where rating[i] > rating[j]
            right_smaller = 0   # count of k > j where rating[k] < rating[j]
            right_greater = 0   # count of k > j where rating[k] > rating[j]
            
            # check left side of j
            for i in range(j):
                if rating[i] < rating[j]:
                    left_smaller += 1
                elif rating[i] > rating[j]:
                    left_greater += 1
            
            # check right side of j
            for k in range(j + 1, n):
                if rating[k] > rating[j]:
                    right_greater += 1
                elif rating[k] < rating[j]:
                    right_smaller += 1
            
            # increasing teams + decreasing teams
            count += left_smaller * right_greater
            count += left_greater * right_smaller
        
        return count
```

```python
s = Solution()
print(s.numTeams([2,5,3,4,1]))  
```
